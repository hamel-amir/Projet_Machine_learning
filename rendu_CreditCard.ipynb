**Détection et traitements des valeurs aberrantes pour chaque variable**

```python
def find_outliers_IQR(df):
   q1=df.quantile(0.25)

   q3=df.quantile(0.75)

   IQR=q3-q1

   outliers = df[((df<(q1-1.5*IQR)) | (df>(q3+1.5*IQR)))]

   return outliers

# passer une variable en parametre de la fonction
outliers=find_outliers_IQR(df['V10'])
print("number of outliers: "+ str(len(outliers)))

print("max outlier value: "+ str(outliers.max()))

print("min outlier value: "+ str(outliers.min()))
```

Fonction qui construit la matrice de confusion

```python
import itertools
import numpy as np

# Create a confusion matrix
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print(" ")
    else:
        print(' ')

    print(cm)

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title, fontsize=14)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```

**Arbre de décision**

```python
from sklearn import tree
from sklearn import metrics
clf = tree.DecisionTreeClassifier()
# fit model
clf = clf.fit(X_train, y_train)
# predict model
y_pred=clf.predict(X_test)
y_pred
```

**les métriques d'évaluation de performances des modeles** 

```python
metrics.confusion_matrix(y_test,y_pred)
import scikitplot as skplt
skplt.metrics.plot_confusion_matrix(y_test, y_pred)
```

D'apres la matrice de confusion on remarque que : 

* TP ( true positive ) , le nombre d'instance positives (fraud) qui ont été predites positive 

* TN ( true negative ), le nombre d'instances negatives ( classe=0 non-fraud) qui ont été predites negatives 

* FN (false negative), le nombre d'instances negatives( classe=0) qui ont été predites positives (incorrecte).

* FN (false positive ), le nombre d'instances positive( classe=1) qui ont été predites negative.


```python
def metrics_evaluation(y_test,y_pred):
  
  print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
  print("--------------------------------------------------")
  print("Recall:",metrics.recall_score(y_test, y_pred))
  print("--------------------------------------------------")
  print("precision:",metrics.precision_score(y_test, y_pred))
  print("--------------------------------------------------")
  print("f1_score:",metrics.f1_score(y_test, y_pred))
  print("--------------------------------------------------")
  
metrics_evaluation(y_test,y_pred)
```

D'aprés les résultats des mesures de performances qu'on a obtenu , on peut constater que nore modele est bon en terme  : 

* accuracy ( le taux d'observation predites correctement ) = 0.999

*  Recall ( la proportion d'observation positive qui'ont été predit correctement ) = 0.7972

* precision_score : la proportion d’observation positifs prédit qui sont réellement positif = 0.737

* f1_score:  est la moyenne du rappel et de la précision , un score plus élevé étant un meilleur modèle = 0.766
mais  ça ne représente pas la réalité de notre dataset car les classes sont désequilibrés et ya une sur presentation de la classe 0 par rapport a la classe 1 et cela va entrainer des impacts sur la performance du modele predictif .


```python
import matplotlib.pyplot as plt
def curbe_ROC(y_test,y_pred):
  fpr_RF, tpr_RF, thresholds_RF = metrics.roc_curve(y_test,y_pred)
  
  # AUC
  auc_RF = metrics.roc_auc_score(y_test,y_pred)

  plt.plot(fpr_RF, tpr_RF,'k-',label = f'AUC =({auc_RF})')
  # plt.plot([0,1],[0,1],'r-',label='random model')
  # plt.plot([0,0,1,1],[0,1,1,1],'g-',label='perfect model')
  plt.title('ROC Curve', fontsize=16)
  plt.legend()
  plt.xlabel('False Positive Rate')
  plt.ylabel('True Positive Rate')
  plt.show()
  
curbe_ROC(y_test,y_pred)
```

**Définition des variables (features) les plus importantes**

c'est une téchnique permet de determiner les variables les plus pertinenetes (en fonction des scores ) et celles qui sont faible afin d'améliorer les performances du modèle.

```python
def features_importance(X_train,feature_importance,modal_name):

    # Créer une liste de tuples (nom de la caractéristique, importance) pour chaque caractéristique
    feature_importance_tuples = [(X_train.columns[i], importance) for i, importance in enumerate(feature_importance)]

    # Trier la liste en fonction de l'importance (du plus grand au plus petit)
    feature_importance_tuples.sort(key=lambda x: x[1], reverse=True)

    # Afficher l'importance des caractéristiques avec leurs noms triés
    for feature, importance in feature_importance_tuples:
        print(f"Feature '{feature}': Importance = {importance:.4f}")

    # Plotter l'importance des caractéristiques sous forme de graphique à barres

    # Extraire les noms des caractéristiques et leurs importances triées
    feature_names = [feature[0] for feature in feature_importance_tuples]
    importances = [feature[1] for feature in feature_importance_tuples]

    # Plotter les importances des caractéristiques
    plt.bar(range(len(feature_names)), importances)
    plt.xticks(range(len(feature_names)), feature_names, rotation='vertical')
    plt.xlabel('Feature')
    plt.ylabel('Importance')
    plt.title(f'Importance des caractéristiques pour {modal_name}')
    plt.show()
```
