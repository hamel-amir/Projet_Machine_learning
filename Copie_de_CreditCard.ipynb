**Détection et traitements des valeurs aberrantes pour chaque variable**

```python
def find_outliers_IQR(df):
   q1=df.quantile(0.25)

   q3=df.quantile(0.75)

   IQR=q3-q1

   outliers = df[((df<(q1-1.5*IQR)) | (df>(q3+1.5*IQR)))]

   return outliers

# passer une variable en parametre de la fonction
outliers=find_outliers_IQR(df['V10'])
print("number of outliers: "+ str(len(outliers)))

print("max outlier value: "+ str(outliers.max()))

print("min outlier value: "+ str(outliers.min()))
```

**Les modeles de classification**

```python
from sklearn.model_selection import train_test_split
Y=df['Class'] # target
mask=df.columns.shape[0]
features=df.columns[:mask-1]
X=df[features] # features

X_train, X_test, y_train, y_test = train_test_split(X, Y, random_state=1, train_size = .7, stratify=Y)
```

```python
X_train=X_train[['V1', 'V2', 'V3', 'V4', 'V5', 'V6', 'V7', 'V8', 'V9', 'V10',
       'V11', 'V12', 'V13', 'V14', 'V15', 'V16', 'V17', 'V18', 'V19', 'V20',
       'V21', 'V22', 'V23', 'V24', 'V25', 'V26', 'V27', 'V28']]

X_test=X_test[['V1', 'V2', 'V3', 'V4', 'V5', 'V6', 'V7', 'V8', 'V9', 'V10',
       'V11', 'V12', 'V13', 'V14', 'V15', 'V16', 'V17', 'V18', 'V19', 'V20',
       'V21', 'V22', 'V23', 'V24', 'V25', 'V26', 'V27', 'V28']]
```

**Arbre de décision**

```python
from sklearn import tree
from sklearn import metrics
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)
#tree.plot_tree(clf)
y_pred=clf.predict(X_test)
y_pred
```

```python
def metrics_evaluation(y_test,y_pred):
  # Model Accuracy, how often is the classifier correct?
  print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
  print("--------------------------------------------------")
  print("Recall:",metrics.recall_score(y_test, y_pred))
  print("--------------------------------------------------")
  print("precision:",metrics.precision_score(y_test, y_pred))
  print("--------------------------------------------------")
  print("f1_score:",metrics.f1_score(y_test, y_pred))
  print("--------------------------------------------------")
  



metrics_evaluation(y_test,y_pred)
```

```python
metrics.confusion_matrix(y_test,y_pred)
import scikitplot as skplt
skplt.metrics.plot_confusion_matrix(y_test, y_pred)
```

```python
def curbe_ROC(y_test,y_pred):
  fpr_RF, tpr_RF, thresholds_RF = metrics.roc_curve(y_test,y_pred)
  import matplotlib.pyplot as plt
  # AUC
  auc_RF = metrics.roc_auc_score(y_test,y_pred)

  plt.plot(fpr_RF, tpr_RF,'k-',label = f'tree model: AUC =({auc_RF})')
  # plt.plot([0,1],[0,1],'r-',label='random model')
  # plt.plot([0,0,1,1],[0,1,1,1],'g-',label='perfect model')
  plt.legend()
  plt.xlabel('False Positive Rate')
  plt.ylabel('True Positive Rate')
  plt.show()
  
curbe_ROC(y_test,y_pred)
```

**Comparaison de différente approches de SVM en fonction de la courbe ROC et les métriques d'évaluation**


```python
def curbe_ROC_SVM():
  fpr_RF, tpr_RF, thresholds_RF = metrics.roc_curve(y_test,svm_pred)
  sigmoid_fpr_RF, sigmoid_tpr_RF, sigmoid_thresholds_RF = metrics.roc_curve(y_test,sigmoid_pred)
  linear_fpr_RF, linear_tpr_RF, linear_thresholds_RF = metrics.roc_curve(y_test,linear_pred)
  poly_fpr_RF, poly_tpr_RF, poly_thresholds_RF = metrics.roc_curve(y_test,poly_pred)
  import matplotlib.pyplot as plt
  # AUC
  auc_RF = metrics.roc_auc_score(y_test,y_pred)
  sigmoid_auc_RF = metrics.roc_auc_score(y_test,sigmoid_pred)
  linear_auc_RF = metrics.roc_auc_score(y_test,linear_pred)
  poly_auc_RF = metrics.roc_auc_score(y_test,poly_pred)

  plt.plot(fpr_RF, tpr_RF,'b-',label = f'SVM sans kernel: AUC =({auc_RF})')
  plt.plot(sigmoid_fpr_RF,sigmoid_tpr_RF,'k-',label = f' SVM (sigmoid): AUC =({sigmoid_auc_RF})')
  plt.plot(linear_fpr_RF,linear_tpr_RF,'g-',label = f'SVM (linear): AUC =({linear_auc_RF})')
  poly_fpr_RF, poly_tpr_RF, poly_thresholds_RF = metrics.roc_curve(y_test,poly_pred)
  plt.plot(poly_fpr_RF,poly_tpr_RF,'r-',label = f'SVM (poly): AUC =({poly_auc_RF})')
  # plt.plot([0,1],[0,1],'r-',label='random model')
  # plt.plot([0,0,1,1],[0,1,1,1],'g-',label='perfect model')
  plt.legend()
  plt.xlabel('False Positive Rate')
  plt.ylabel('True Positive Rate')
  plt.show()


curbe_ROC_SVM()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
params = {
    "criterion":("gini", "entropy"), 
    "max_depth":list(range(2, 4)), 
    "min_samples_split":[2, 3], 
    "min_samples_leaf":list(range(1, 2)), 

}
tree_clf=DecisionTreeClassifier(random_state=42)
grid_search_cv = GridSearchCV(
    tree_clf,
     params,
     n_jobs=1,
     verbose=1,
     cv=2)
grid_search_cv.fit(X_train, y_train)
```
