**Détection et traitements des valeurs aberrantes**

```python
def find_outliers_IQR(df):
   q1=df.quantile(0.25)

   q3=df.quantile(0.75)

   IQR=q3-q1

   outliers = df[((df<(q1-1.5*IQR)) | (df>(q3+1.5*IQR)))]

   return outliers

outliers=find_outliers_IQR(df['V1'])
print("number of outliers: "+ str(len(outliers)))

print("max outlier value: "+ str(outliers.max()))

print("min outlier value: "+ str(outliers.min()))
```

**Rééquilibrage des classes**

```python
from collections import Counter
from imblearn.over_sampling import RandomOverSampler
Y=df['Class'] # target
mask=df.columns.shape[0]
features=df.columns[:mask-1]
X=df[features] # features
#print(Counter(Y))
# define oversampling strategy
oversample = RandomOverSampler(sampling_strategy='minority')
# fit and apply the transform
X_over, y_over = oversample.fit_resample(X, Y)
# summarize class distribution
#print(Counter(y_over))
X_over['Class']=y_over
df=X_over.copy()
sns.countplot(x='Class', data=df, palette=colors)
```

**Les modeles de classification**

```python
from sklearn.model_selection import train_test_split
Y=df['Class'] # target
mask=df.columns.shape[0]
features=df.columns[:mask-1]
X=df[features] # features

X_train, X_test, y_train, y_test = train_test_split(X, Y, random_state=1, train_size = .7, stratify=Y)
```

**Arbre de décision**

```python
from sklearn import tree
from sklearn import metrics
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)
#tree.plot_tree(clf)
y_pred=clf.predict(X_test)
y_pred
```

```python
def metrics_evaluation(y_test,y_pred):
  # Model Accuracy, how often is the classifier correct?
  print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
  print("--------------------------------------------------")
  print("Recall:",metrics.recall_score(y_test, y_pred))
  print("--------------------------------------------------")
  print("precision:",metrics.precision_score(y_test, y_pred))
  print("--------------------------------------------------")
  print("f1_score:",metrics.f1_score(y_test, y_pred))
  print("--------------------------------------------------")
  



metrics_evaluation(y_test,y_pred)
```

```python
metrics.confusion_matrix(y_test,y_pred)
import scikitplot as skplt
skplt.metrics.plot_confusion_matrix(y_test, y_pred)
```

```python
def curbe_ROC(y_test,y_pred):
  fpr_RF, tpr_RF, thresholds_RF = metrics.roc_curve(y_test,y_pred)
  import matplotlib.pyplot as plt
  # AUC
  auc_RF = metrics.roc_auc_score(y_test,y_pred)

  plt.plot(fpr_RF, tpr_RF,'k-',label = f'tree model: AUC =({auc_RF})')
  plt.plot([0,1],[0,1],'r-',label='random model')
  plt.plot([0,0,1,1],[0,1,1,1],'g-',label='perfect model')
  plt.legend()
  plt.xlabel('False Positive Rate')
  plt.ylabel('True Positive Rate')
  plt.show()
  
curbe_ROC(y_test,y_pred)
```
