```python
import itertools
import numpy as np

# Create a confusion matrix
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print(" ")
    else:
        print(' ')

    print(cm)

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title, fontsize=14)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```

```python
def metrics_evaluation(y_test,y_pred):
  # Model Accuracy, how often is the classifier correct?
  print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
  print("--------------------------------------------------")
  print("Recall:",metrics.recall_score(y_test, y_pred))
  print("--------------------------------------------------")
  print("precision:",metrics.precision_score(y_test, y_pred))
  print("--------------------------------------------------")
  print("f1_score:",metrics.f1_score(y_test, y_pred))
  print("--------------------------------------------------")
  
metrics_evaluation(y_test,y_pred)
```

D'aprés les résultats des mesures de performances qu'on a obtenu , on peut constater que nore modele est bon en terme  : 

* accuracy ( le taux d'observation predites correctement ) = 0.999

*  Recall ( la proportion d'observation positive qui'ont été predit correctement ) = 0.7972

* precision_score : la proportion d’observation positifs prédit qui sont réellement positif = 0.737

* f1_score:  est la moyenne du rappel et de la précision , un score plus élevé étant un meilleur modèle = 0.766
mais  ça ne représente pas la réalité de notre dataset car les classes sont désequilibrés et ya une sur presentation de la classe 0 par rapport a la classe 1 et cela va entrainer des impacts sur la performance du modele predictif .


```python
def curbe_ROC(y_test,y_pred):
  fpr_RF, tpr_RF, thresholds_RF = metrics.roc_curve(y_test,y_pred)
  import matplotlib.pyplot as plt
  # AUC
  auc_RF = metrics.roc_auc_score(y_test,y_pred)

  plt.plot(fpr_RF, tpr_RF,'k-',label = f'AUC =({auc_RF})')
  # plt.plot([0,1],[0,1],'r-',label='random model')
  # plt.plot([0,0,1,1],[0,1,1,1],'g-',label='perfect model')
  plt.legend()
  plt.xlabel('False Positive Rate')
  plt.ylabel('True Positive Rate')
  plt.show()
  
curbe_ROC(y_test,y_pred)
```
